## Lo primero es importar todas las librearías que nos pueden servir

# Ahora voy a cargar el cubo como un SpectralCube

```python
cube2 = SpectralCube.read('VPs_MUSE/sdss1524.fits',hdu=1)
```

# Los momento son faciles de hacer. 

```python
moment_0 = cube2.moment(order=0)
moment_1 = cube2.moment(order=1)
moment_2 = cube2.moment(order=2)
```

```python
moment_0.quicklook()
```

```python
moment_1.quicklook()
```

```python
moment_2.quicklook()
```

# Ahora voy a buscar la referencia para los momentos

```python
cube[:,:,:].sum(axis=(1,2)).plot(unit=u.angstrom,lmin=5000,lmax=5300)
```

```python
cube[:,:,:].sum(axis=(1,2)).plot(unit=u.angstrom,lmin=5170,lmax=5210)
```

```python
OIII = cube[:,:,:].sum(axis=(1,2)).gauss_fit(unit=u.angstrom,lmin=5170,lmax=5210, plot=True)
OIII.print_param()
```

```python
OIII_cube = cube2.with_spectral_unit(u.km/u.s,velocity_convention='optical',rest_value=5191.71*u.AA)
```

```python
OIII_subcube = OIII_cube.spectral_slab(-600*u.km/u.s,600*u.km/u.s)
```

```python
moment_1 = OIII_subcube.moment(order=1)
```

```python
moment_1.quicklook('moment_1.png')
```

```python
plt.imshow(moment_1.value,origin='lower',vmin=-300,vmax=300,cmap='coolwarm')
plt.colorbar()
```

```python
Cube_MUSE = fits.open('VPs_MUSE/sdss1524.fits')
cube2_sn = SpectralCube(data=Cube_MUSE[1].data/np.sqrt(Cube_MUSE[2].data), wcs=cube2.wcs)
```

```python
cube2_masked = cube2.with_mask(cube2_sn>2.5)
```

```python
OIII_cube_masked = cube2_masked.with_spectral_unit(u.km/u.s,velocity_convention='optical',rest_value=5191.71*u.AA)
OIII_subcube_masked = OIII_cube_masked.spectral_slab(-600*u.km/u.s,600*u.km/u.s)
```

```python
moment_1_masked = OIII_subcube_masked.moment(order=1)
plt.imshow(moment_1_masked.value,origin='lower',vmin=-300,vmax=300,cmap='coolwarm')
plt.colorbar()
```

```python
moment_0_masked = OIII_subcube_masked.moment(order=0)
plt.imshow(moment_0_masked.value,origin='lower',cmap='inferno',vmax=1e5)
plt.colorbar()
```

```python
moment_2_masked = OIII_subcube_masked.moment(order=2)
plt.imshow(np.sqrt(moment_2_masked.value),origin='lower',cmap='inferno',vmin=0,vmax=400)
plt.colorbar()
```
