```python
line1 = cube[450:490,150,160]
# line1.plot()
line1 = line1 - 288
line1.plot()
line1.gauss_fit(unit=u.angstrom,lmin=5168,lmax=5210).print_param()


# lineunit = (line1.unit.to(u.Jy, equivalencies=u.dimensionless_angles())*u.Jy).to(u.erg / u.cm**2 / u.s / u.micron,
#                 equivalencies=u.spectral_density(1.529 * u.micron))

fline = (line1.sum()*line1.unit) * (line1.get_step(unit=line1.wave.unit)*line1.wave.unit)
fline
```

```python
from scipy.integrate import simpson
```

```python
Flux = simpson(line1.data,x=line1.wave.coord())*line1.unit*line1.wave.unit
```

```python
from astropy.cosmology import WMAP9 as cosmo
```

```python
z = 5192.05/5008-1
dL = cosmo.luminosity_distance(z)
L = 4*np.pi*dL*dL*(Flux).to(u.erg / u.cm**2 / u.s)
L.to(u.erg/u.s),L.to(u.Lsun)
```
