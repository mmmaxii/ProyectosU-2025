# Calculo de errores

## Funcion para calcular el scaling

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy.modeling import models, fitting

def calculate_variance_scaling(cube, var_cube, wavelength_range=None):
    """
    Calcula el factor de escalamiento necesario para la varianza
    
    Parameters:
    -----------
    cube : array 3D
        Cubo de datos (lambda, y, x)
    var_cube : array 3D  
        Cubo de varianza (lambda, y, x)
    mask_sky : array 2D bool, optional
        Máscara de zonas de cielo (True = cielo)
    wavelength_range : tuple, optional
        Rango de longitudes de onda a usar (lmin, lmax)
    
    Returns:
    --------
    scaling_factors : array 1D
        Factores de escalamiento por canal espectral
    """
    
    # Si no se proporciona máscara de cielo, usar todo el campo
    
    
    # Si no se especifica rango espectral, usar todo
    if wavelength_range is None:
        wavelength_range = (0, cube.shape[0])
    
    lmin, lmax = wavelength_range
    scaling_factors = []
    
    # Calcular para cada canal espectral en el rango
    for i in range(lmin, lmax):
        # Extraer imagen 2D para este canal
        data_slice = cube[i, :, :].data
        var_slice = var_cube[i, :, :]
        # Aplicar máscara de cielo
        sky_data = data_slice
        sky_var = var_slice

        # Calcular RMS observado (ruido real)
        rms_observed = np.sqrt((sky_data**2).mean(axis=(0,1)))
        
        # Calcular RMS teórico (promedio de sqrt(varianza))
        rms_theoretical = np.sqrt(sky_var.mean(axis=(0,1)))
        
        # Factor de escalamiento (debería ser ~1 si está bien)
        if rms_theoretical > 0:
            scaling_factor = rms_observed / rms_theoretical
        else:
            scaling_factor = 1.0
            
        scaling_factors.append(scaling_factor)
    
    return np.array(scaling_factors)
```

## Funcion para crear la mask del cielo

```python
import numpy as np

y_min, y_max = 208, 269  
x_min, x_max = 49, 115

cube_sky_region = cube[:, y_min:y_max, x_min:x_max] ## ESTO ES LO IMPORTANTE, YA EXTRAIMOS UNA PARTE DEL CIELO QUE PRESENTA ESTO.
print(np.sqrt(cube_sky_region[100,:,:].var.mean(axis=(0,1)))) # Teorico
print(np.sqrt((cube_sky_region[100,:,:].data**2).mean(axis=(0,1)))) # RMS
print(np.std(cube_sky_region[100,:,:].data)) # std

valores_de_escalamiento = calculate_variance_scaling(cube_sky_region, cube_sky_region.var)
```

```python
# Procedimiento final

for i in range(len(valores_de_escalamiento)):
    cube.var[i, :, :] = cube.var[i, :, :] * (valores_de_escalamiento[i] ** 2)
```

# Hoy vamos a ver como hacer bins de Voronoi para mejor analizar los datos.

```python
cube[450:490,:,:].max(axis=0).plot()
```

```python
index_max = np.argmax(cube[450:490,:,:].data,axis=0)
plt.imshow(index_max,origin='lower',cmap='viridis')
plt.colorbar()
```

```python
Error = np.sqrt(cube[450:490,:,:].var)
ErrorM8 = np.zeros_like(np.argmax(cube[450:490,:,:].data,axis=0))
ErrorM8 = ErrorM8.astype(float)
contador = 0
for i in range(len(index_max)):
    for j in range(len(index_max[i])):
        aux = index_max[i][j]
        if Error.mask[aux][i][j]:
            continue
        else:
            ErrorM8[i][j] = Error[aux][i][j]
print(contador,51*53)
ErrorM8 = ErrorM8.astype(float)
ErrorM8[ErrorM8==0] = np.nan
```

```python
plt.imshow(ErrorM8,origin='lower',vmax=10)
plt.colorbar()
```

```python
import vorbin
from vorbin.voronoi_2d_binning import voronoi_2d_binning
```

```python
signal = cube[450:490,:,:].max(axis=0).data
noise = ErrorM8
sn = signal/noise

x,y = np.nonzero(sn>5)
```

```python
plt.imshow(sn,origin='lower',vmax=20)
plt.colorbar()
```

Ahora que corregi el ruido lo que paso es que tuve que bajar el valor de 50 a un valor mas bajo pq ahora habian muchos menos bins de voronoi, vamos por buen camino, sera esencial ponerlo ene l informe.

```python
binNum, x_gen, y_gen, x_bar, y_bar, sn, nPixels, scale = voronoi_2d_binning(x, y, signal[x,y], noise[x,y], 28, plot=1, quiet=1,pixelsize=0.2)
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap

plt.figure(figsize=(12, 10))  # Tamaño grande como en papers profesionales

# CALCULAR EL CENTRO Y CENTRAR COORDENADAS
x_center = 215  # Usando el centro que ya estabas usando en xlim
y_center = 225  # Usando el centro que ya estabas usando en ylim

# Centrar las coordenadas (0,0 en el centro de la imagen)
x_centered = x - x_center
y_centered = y - y_center
x_gen_centered = x_gen - x_center
y_gen_centered = y_gen - y_center

np.random.seed(42)  # Para reproducibilidad
unique_bins = np.unique(binNum)
color_mapping = np.random.permutation(len(unique_bins))

# Mapear cada bin a un color mezclado
bin_to_color = {bin_id: color_mapping[i] for i, bin_id in enumerate(unique_bins)}
mixed_colors = np.array([bin_to_color[bin_id] for bin_id in binNum])

# Scatter con coordenadas CENTRADAS
plt.scatter(x_centered, y_centered, c=mixed_colors, cmap='tab20', s=25, alpha=0.8, marker="s")
plt.scatter(x_gen_centered, y_gen_centered, c='b', s=1, alpha=0.9, marker='o', edgecolors='black', linewidth=1)

plt.xlabel('arcsec', fontsize=15)
plt.ylabel('arcsec', fontsize=15)
plt.title('Teselación de Voronoi - NGC 5972', fontsize=16, fontweight='bold')

# Asegurar proporción cuadrada como en el paper
plt.gca().set_aspect('equal')

# Configurar ejes como regla métrica
plt.tick_params(which='both', direction='in', top=True, right=True, 
               bottom=True, left=True, labelsize=10)

# LÍMITES CENTRADOS (valores negativos y positivos)
plt.xlim(-60, 60)  # En lugar de 215-60, 215+60
plt.ylim(-60, 60)  # En lugar de 225-60, 225+60



# Mejorar ticks y estilo general
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.grid(True, alpha=0.2)



plt.tight_layout()
plt.savefig("Teselacion de voronoi")
plt.show()

# Mostrar información del centrado
print(f"Centro usado: X={x_center}, Y={y_center}")
print(f"Rango X: [{np.min(x_centered):.1f}, {np.max(x_centered):.1f}] arcsec")
print(f"Rango Y: [{np.min(y_centered):.1f}, {np.max(y_centered):.1f}] arcsec")
```
