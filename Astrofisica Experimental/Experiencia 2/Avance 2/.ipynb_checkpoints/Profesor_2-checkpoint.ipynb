# Hoy vamos a ver como hacer bins de Voronoi para mejor analizar los datos.

```python
cube[450:490,:,:].max(axis=0).plot()
```

```python
index_max = np.argmax(cube[450:490,:,:].data,axis=0)
plt.imshow(index_max,origin='lower',cmap='viridis')
plt.colorbar()
```

```python
Error = np.sqrt(cube[450:490,:,:].var)
ErrorM8 = np.zeros_like(np.argmax(cube[450:490,:,:].data,axis=0))
ErrorM8 = ErrorM8.astype(float)
contador = 0
for i in range(len(index_max)):
    for j in range(len(index_max[i])):
        aux = index_max[i][j]
        if Error.mask[aux][i][j]:
            continue
        else:
            ErrorM8[i][j] = Error[aux][i][j]
print(contador,51*53)
ErrorM8 = ErrorM8.astype(float)
ErrorM8[ErrorM8==0] = np.nan
```

```python
plt.imshow(ErrorM8,origin='lower',vmax=10)
plt.colorbar()
```

```python
import vorbin
from vorbin.voronoi_2d_binning import voronoi_2d_binning
```

```python
signal = cube[450:490,:,:].max(axis=0).data
noise = ErrorM8
sn = signal/noise

x,y = np.nonzero(sn>5)
```

```python
plt.imshow(sn,origin='lower',vmax=20)
plt.colorbar()
```

```python
binNum, x_gen, y_gen, x_bar, y_bar, sn, nPixels, scale = voronoi_2d_binning(x, y, signal[x,y], noise[x,y], 100, plot=1, quiet=1,pixelsize=0.2)



```

```python
bin2use = np.argmin(sn)
print(x[binNum==bin2use],y[binNum==bin2use])
cube2 = cube.copy()
cube2.mask = np.ones_like(cube2.mask)*True
cube2.mask[:,x[binNum==bin2use],y[binNum==bin2use]] = np.zeros_like(cube2.mask[:,x[binNum==bin2use],y[binNum==bin2use]])*True
cube2.sum(axis=(1,2)).plot(unit=u.angstrom,lmin=5170,lmax=5210,snr=True)
```

```python
binNum, x_gen, y_gen, x_bar, y_bar, sn, nPixels, scale = voronoi_2d_binning(x, y, signal[x,y], noise[x,y], 50, plot=1, quiet=1,pixelsize=0.2)
```

```python
bin2use = np.argmin(sn)
bin2use = 7
print(x[binNum==bin2use],y[binNum==bin2use])
cube2 = cube.copy()
cube2.mask = np.ones_like(cube2.mask)*True
cube2.mask[:,x[binNum==bin2use],y[binNum==bin2use]] = np.zeros_like(cube2.mask[:,x[binNum==bin2use],y[binNum==bin2use]])*True
cube2.sum(axis=(1,2)).plot(unit=u.angstrom,lmin=5170,lmax=5210,snr=True)
```
